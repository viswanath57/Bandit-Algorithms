```r
library("plyr")
library("ggplot2")

results <- read.csv("D:/BanditsSimulationDataSets/epsilon-greedy-with-multiple-epsilons.tsv", header = FALSE, sep='\t')
names(results) <- c("Epsilon", "Sim", "T", "ChosenArm", "Reward", "CumulativeReward")
results <- transform(results, Epsilon = factor(Epsilon))

# Plot average reward as a function of time.
stats <- ddply(results,
               c("Epsilon", "T"),
               function (df) {mean(df$Reward)})
ggplot(stats, aes(x = T, y = V1, group = Epsilon, color = Epsilon)) +
  geom_line() +
  ylim(0, 1) +
  xlab("Time") +
  ylab("Average Reward") +
  ggtitle("Performance of the Epsilon Greedy Algorithm with 5 arms with estimated success rates :  [0.1, 0.1, 0.1, 0.1, 0.9]")
ggsave("plots/standard_epsilon_greedy_average_reward.pdf")
```

```r
# Plot average reward as a function of time.
stats <- ddply(results,
               c("Epsilon", "T"),
               function (df) {mean(df$ChosenArm == 2)})
ggplot(stats, aes(x = T, y = V1, group = Epsilon, color = Epsilon)) +
  geom_line() +
  ylim(0, 1) +
  xlab("Time") +
  ylab("Accuracy in choosing right arm") +
  ggtitle("Performance of the Epsilon Greedy Algorithm with 5 arms with estimated success rates :  [0.1, 0.1, 0.1, 0.1, 0.9]")
ggsave("plots/standard_epsilon_greedy_accuracy_in_choosing_right_arm.pdf")
```

```r
# Plot average reward as a function of time.
stats <- ddply(results,
               c("Epsilon", "T"),
               function (df) {mean(df$CumulativeReward)})

ggplot(stats, aes(x = T, y = V1, group = Epsilon, color = Epsilon)) +
  geom_line() +
  xlab("Time") +
  ylab("Cummulative Reward") +
  ggtitle("Performance of the Epsilon Greedy Algorithm with 5 arms with estimated success rates :  [0.1, 0.1, 0.1, 0.1, 0.9]")
ggsave("plots/standard_epsilon_greedy_cummulative_reward.pdf")
```

```r
ucb <- read.table("ucb1_esults.tsv", sep = '\t')
ucb$algo <- rep("ucb", dim(ucb)[1] )
names(ucb) <- c("Sim", "T", "ChosenArm", "Reward", "CumulativeReward", "algo")

annealing_softmax <- read.table("annealing_softmax_results.tsv", sep = '\t')
annealing_softmax$algo <- rep ( "softmax", dim(annealing_softmax)[1] )
names(annealing_softmax) <- c("Sim", "T", "ChosenArm", "Reward", "CumulativeReward", "algo")

annealing_eg <- read.table("annealing_epsilon_greedy_results.tsv", sep = '\t')
annealing_eg$algo <- rep("annealing_eg", dim(ucb)[1] )
names(annealing_eg) <- c("Sim", "T", "ChosenArm", "Reward", "CumulativeReward", "algo")

df <- rbind( ucb , annealing_softmax, annealing_eg)
```

```r
# Plot average reward as a function of time.
stats <- ddply(df,
               c("algo", "T"),
               function (df) {mean(df$Reward)})
ggplot(stats, aes(x = T, y = V1, group = algo, color = algo)) +
  geom_line() +
  ylim(0, 1) +
  xlab("Time") +
  ylab("Average Reward") +
  ggtitle("Performance of the Multiarm Bandit Algorithms")
```

```r
# Plot average reward as a function of time.
stats <- ddply(df,
               c("algo", "T"),
               function (df) {mean(df$CumulativeReward)})
ggplot(stats, aes(x = T, y = V1, group = algo, color = algo)) +
  geom_line() +
  xlab("Time") +
  ylab("Cumulative  Reward") +
  ggtitle("Performance of MultiArm Bandit Algorithms")

```

```r
# Plot average reward as a function of time.
stats <- ddply(df,
               c("algo", "T"),
               function (df) {mean(df$ChosenArm == 2)})
ggplot(stats, aes(x = T, y = V1, group = algo, color = algo)) +
  geom_line() +
  ylim(0, 1) +
  xlab("Time") +
  ylab("Accuracy") +
  ggtitle("Performance of the Multiarm Bandit Algorithms")
```
